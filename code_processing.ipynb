```python
import pandas as pd
import seaborn as sns
```

```python
chunk_size = 10 ** 6  # Number of lines to read at a time
chunks = []

for chunk in pd.read_csv('datasets/weekly_housing_market_data_most_recent.tsv000', delimiter='\t', chunksize=chunk_size):
    # Process each chunk here
    # For example, you might filter rows or columns and append the processed chunk to a list
    chunks.append(chunk)

# Concatenate all chunks into a single DataFrame
redfin_data = pd.concat(chunks, ignore_index=True)
```

```python
redfin_data.filter(like="price")
redfin_data = redfin_data.dropna(axis=1, how='all')
```
